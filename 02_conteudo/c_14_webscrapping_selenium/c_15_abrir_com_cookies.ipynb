### Abrindo o navegador

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())

options = webdriver.ChromeOptions()

# Para localizar onde estão armazenados os dados do perfil, basta digitar chrome://version/
# Dentro do link, procurar por caminho do perfil

options.add_argument(r'user-data-dir= C:\Users\Usuario\AppData\Local\AVAST Software\Browser\User Data\Default')
#Não é recomendado abrir desta forma. Pois podem acontecer muitas incompatibilidades

# Ao definir um perfil que ainda não exista, este perfil é criado automaticamente na página indicada
# options.add_argument(r'user-data-dir= C:\Users\Usuario\AppData\Local\AVAST Software\Browser\User Data\Profile Selenium')
# Depois de rodar a primeira vez. somente busca o perfil já criado.
navegador = webdriver.Chrome(service=servico, options=options)
```

```python
navegador.get("https://web.whatsapp.com")
```
