### Abrindo o navegador

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)


# OBSERVAÇÃO IMPORTANTE - O PRINT SEMPRE IMPRIME EXATAMENTE OQUE O NAVEGADOR ESTÁ MOSTRANDO
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\Pagina Hashtag.html"
navegador.get(arquivo)
```

### Tirando print da tela inteira

```python
# Salva a foto com o arquivo com o nome informado entre parênteses
navegador.save_screenshot("print.png")
# navegador.save_screenshot("../print.jpg") - altera o formato e local do arquivo - conceito de caminhos
```

### Tirando print de parte da tela

```python
#! pip install pillow
#! pip install opencv-python

# Se necessário , instalar as libs adicionais
#! pip install opencv-python-headless
#! pip install opencv-contrib-python
# Utilizar pillow ou opencv

# PIL é utilizado para chamar o pillow
from PIL import Image

# Utiliza o print da imagem tirado anteriormente
imagem = Image.open("print.png")

# Pega a posição e tamanho do elemento desejado, podendo tirar print somente do elemento, por exemplo.
# Desconsidera os headers do site, inicia na página efetivamente.
elemento = navegador.find_element(By.ID, 'header')
posicao = elemento.location
# print(posicao)
tamanho = elemento.size
# print(tamanho)
# tamanho e posição retornam tuplas com as variáveis e valores

x_inicial = posicao["x"] * 2.5
y_inicial = posicao["y"] * 2.5
x_final = (x_inicial + tamanho["width"]) * 2.5
y_final = (y_inicial + tamanho["height"]) * 2.5

# 2,5 refere-se às configurações do computador
# Para verificar: sistema > tela > escala e layout > escala
# Ver imagem abaixo para dúvidas
# No caso da imagem, o valor deveria ser substituido por 1, pois 2.5 refere-se a 250%

# crop é utilizado para cortar a imagem.
# Receber como parâmetro uma TUPLA com os quatro cantos que devem servir como limite para o corte
# Os eixos iniciam em (0,0), no canto superior esquerdo, ou seja, o x aumenta para a direita e y para baixo
imagem = imagem.crop((x_inicial, y_inicial, x_final, y_final))
imagem.save("print_pedaco.png")
```
